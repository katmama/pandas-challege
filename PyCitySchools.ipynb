```python
import pandas as pd
import os
```

```python
# File to Load (Remember to Change These)
school_data_to_load = os.path.join("Downloads/schools_complete.csv")
student_data_to_load = os.path.join("Downloads/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
```

```python
#perform calc
school_count = len(school_data_complete_df["school_name"].unique())
total_students = school_data_complete_df["student_name"].count()
avg_budget = school_data_complete_df.groupby('school_name')['budget'].mean()
total_budget = avg_budget.sum()
avg_math_score = school_data_complete_df["math_score"].mean()
avg_read_score = school_data_complete_df["reading_score"].mean()
passing_math = school_data_complete_df[school_data_complete_df.math_score >= 70]
total_math = passing_math["student_name"].count()
math_pass = (total_math/total_students)*100
total_read = school_data_complete_df[school_data_complete_df.reading_score >= 70]["student_name"].count()
read_pass = (total_read/total_students)*100
total_pass = school_data_complete_df[(school_data_complete_df.reading_score >= 70) & (school_data_complete_df.math_score >= 70)]
combined_count = total_pass["student_name"].count()
combined_pass = (combined_count/total_students)*100
```

```python
# Place all of the data found into a summary DataFrame
district_summary_df = pd.DataFrame({"Total Schools": [school_count],
                              "Total Students": total_students,
                              "Total Budget": total_budget,
                              "Average Math Score": avg_math_score,
                              "Average Reading Score": avg_read_score,
                              "% Passing Math": math_pass,
                              "% Passing Reading": read_pass,
                              "% Overall Passing": combined_pass})
district_summary_df
```

```python
#format data

district_summary_df["Total Budget"] = district_summary_df["Total Budget"].astype(float).map("${:,.2f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].astype(float).map("{:.2f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].astype(float).map("{:.2f}".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].astype(float).map("{:.2f}%".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].astype(float).map("{:.2f}%".format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].astype(float).map("{:.2f}%".format)
```

```python
#perform calc
avg_budget = school_data_complete_df.groupby('school_name')['budget'].mean()
students_per_school = school_data_complete_df.groupby('school_name')['student_name'].count()
per_student_budget = avg_budget/students_per_school
sch_avg_math_score = school_data_complete_df.groupby('school_name')['math_score'].mean()
sch_avg_read_score = school_data_complete_df.groupby('school_name')['reading_score'].mean()
math_pass = school_data_complete_df.groupby('school_name').apply(lambda school_data_complete_df: sum(school_data_complete_df.math_score >= 70))
read_pass = school_data_complete_df.groupby('school_name').apply(lambda school_data_complete_df: sum(school_data_complete_df.reading_score >= 70))
sch_mathpass_per = (math_pass/students_per_school)*100
sch_readpass_per = (read_pass/students_per_school)*100
overall = school_data_complete_df[(school_data_complete_df['reading_score'] >= 70) & (school_data_complete_df['math_score'] >= 70)].groupby('school_name')['student_name'].count()/students_per_school
overall_per = overall*100
school_type = school_data_complete_df.groupby('school_name')['type'].unique()
```

```python
school_group = pd.DataFrame({"School Type": school_type,
                               "Total Students": students_per_school,
                               "Total School Budget": avg_budget,
                                "Per Student Budget": per_student_budget,
                                "Average Math Score": sch_avg_math_score,
                                "Average Reading Score": sch_avg_read_score,
                                "% Passing Math": sch_mathpass_per,
                                "% Passing Reading": sch_readpass_per,
                                "% Overall Passing": overall_per})

school_group
```

```python
Scores by School Spending
```

```python
# create spending bins
bins = [0, 584.99, 619.99, 650.999, 999999]
group_name = ['< $585', "$585 - 619", "$620 - 650", "> $650"]
school_data_complete_df['spending_bins'] = pd.cut(school_data_complete_df['budget']/school_data_complete_df['size'], bins, labels = group_name)

#group by spending
by_spend = school_data_complete_df.groupby('spending_bins')


```

```python
#calculations
avg_math = by_spend['math_score'].mean()
avg_read = by_spend['reading_score'].mean()
pass_math = (school_data_complete_df[school_data_complete_df['math_score'] >= 70].groupby('spending_bins')['student_name'].count()/by_spend['student_name'].count())*100
pass_read = (school_data_complete_df[school_data_complete_df['reading_score'] >= 70].groupby('spending_bins')['student_name'].count()/by_spend['student_name'].count())*100
pass_overall = (school_data_complete_df[(school_data_complete_df['reading_score'] >= 70) & (school_data_complete_df['math_score'] >= 70)].groupby('spending_bins')['student_name'].count()/by_spend['student_name'].count())*100


            
```

```python
# df build            
spend_range = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "% Overall Passing": pass_overall
            
})
            

spend_range.index.name = "Spending Ranges (Per Student)"
spend_range = spend_range.reindex(group_name)

spend_range
```

```python
# df build            
scores_by_size = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "% Overall Passing": pass_overall
            
})

scores_by_size.index.name = "School Size"
scores_by_size = scores_by_size.reindex(group_name)

scores_by_size
```
